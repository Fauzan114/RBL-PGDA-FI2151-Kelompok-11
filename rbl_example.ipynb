## abstract
Indonesia Emas 2045 is often touted by the current government. This paper aims to analyze the prediction of the realization of Indonesia Emas 2045 by visualizing several data parameters such as HDI, GDP, and Poverty Rate which will also be compared with several other countries. Data collection was carried out by taking sources from the internet that are guaranteed and data visualization was carried out using scatter plot graphics and linear regression. The results of the analysis answered that 2 out of 3 parameters stated optimism for the occurrence of Indonesia Emas 2045, although the method used was very simple and did not take into account other variables.

## results and dicussion







Based on the existing visualization code, two out of three parameters indicate predictions that align closely with the characteristics of developed countries. The poverty rate and GDP parameters show optimistic results for Indonesia, approaching the characteristics of developed nations such as Switzerland and the United States. However, the HDI parameter presents less promising results, suggesting a less favorable outlook for Indonesia to achieve developed country status.  

From this analysis, it can be optimistically stated that Indonesia Emas 2045 appears to be reasonably achievable. Nonetheless, this assessment does not yet account for other potential factors that may significantly influence the realization of Indonesia Emas 2045.
